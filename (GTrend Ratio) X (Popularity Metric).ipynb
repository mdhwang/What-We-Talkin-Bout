```python
import pandas as pd
```

```python
path = "/Users/matthewhwang/Galvanize/Capstone1/Data/GTrends/Master Lists/Ratios/East_Coast_GTrend_Ratio.csv"
df = pd.read_csv(path)
```

```python
df = df.set_index("Unnamed: 0")
```

```python
EC_spotify = [1581959,517238,4276986,320953,437374,724620,5494448,865204,1664562,9146704,1034099,467370,1416932,4874065,623063,319555,1471420,374281,437300,1241700,378518,511801,954078,534970,759530]
```

```python
weight = EC_spotify
```

```python
names = list(df)
```

```python
for i,each in enumerate(names):
    df[each+" weighted"] = df[each] * weight[i] 
```

```python
df.head()
```

```python
df = df.drop(names,axis=1)
```

```python
df.head()
```

```python
df["total"]=df.sum(axis=1)
```

```python
df["total"] = df["total"].apply(lambda x: int(x))
```
