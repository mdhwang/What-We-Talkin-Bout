```python
import pandas as pd
```

```python
path = "/Users/matthewhwang/Galvanize/Capstone1/Data/GTrends/Master Lists/MidWest_Master.csv"
df = pd.read_csv(path)
```

```python
df = df.set_index("Unnamed: 0")
```

```python
df = df.replace("<1",0)
```

```python
df.head()
```

```python
names = list(df)
```

```python
df = df.astype('int32')
```

```python
df.dtypes
```

```python
for each in names:
    df[each+" frac"] = df[each] / df[each].sum()
```

```python
df = df.drop(names,axis=1)
```
