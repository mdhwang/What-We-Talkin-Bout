```python
import pandas as pd
import matplotlib.pyplot as plt
import glob
```

```python
path = "/Users/matthewhwang/Galvanize/Capstone1/Data/GTrends/Master Lists/Ratios/Spotify Weight Only/*.csv"
for i,each in enumerate(glob.glob(path)):
    name = each.split("Only/")[1].split(".csv")[0]
    df = pd.read_csv(each,names=["date",name])
    if i != 0:
        df_master[name] = df.iloc[:,1]
    else:
        df_master = df
```

```python
df_master.dtypes
```

```python
df_master['date'] = pd.to_datetime(df['date'])
```

```python
df_master.dtypes
```

```python
df_master.head()
```

```python

from matplotlib.pyplot import figure
figure(num=None, figsize=(14, 6), dpi=80, facecolor='w', edgecolor='k')

plt.plot(df_master["date"],df_master["MidWest_Spotify_Totals"],color="r")
plt.plot(df_master["date"],df_master["Dirty_South_Spotify_Totals"],color="b")
plt.plot(df_master["date"],df_master["East_Coast_Spotify_Totals"],color="g")
plt.plot(df_master["date"],df_master["West_Coast_Spotify_Totals"],color="black")
plt.legend()
```

```python
df_cumsum = df_master.iloc[:,1:].cumsum(axis = 0) 
```

```python
df_cumsum["date"] = df_master["date"]
```

```python
figure(num=None, figsize=(14, 6), dpi=80, facecolor='w', edgecolor='k')

plt.plot(df_cumsum["date"],df_cumsum["MidWest_Spotify_Totals"],color="r",label="Mid West",linewidth="3")
plt.plot(df_cumsum["date"],df_cumsum["Dirty_South_Spotify_Totals"],color="b",label="Dirty South",linewidth="3")
plt.plot(df_cumsum["date"],df_cumsum["East_Coast_Spotify_Totals"],color="g",label="East Coast",linewidth="3")
plt.plot(df_cumsum["date"],df_cumsum["West_Coast_Spotify_Totals"],color="black",label="West Coast",linewidth="3")
plt.legend(prop={'size': 16})
plt.xlabel("Years (A.D.)",fontsize="20")
plt.ylabel("Popularity Index",fontsize="20")
plt.title("Regional Rap Popularity in America 2004 - Present",fontsize="30",pad=30)
plt.tick_params(labelsize="large")
```
